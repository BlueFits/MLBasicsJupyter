```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('../data/cost_revenue_clean.csv')
data.describe()
```

```python
x =  DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
def makePlot(x, y, title, xlabel, ylabel):
    plt.scatter(x, y, alpha=0.3)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.ylim(0, 3000000000)
    plt.xlim(0, 450000000)
    plt.show() 
```

```python

makePlot(x, y, "Film Cost vs Global Rev", "Prod Budget", "Worldwide Revenue")
```

Slope Coefficient - You have to understand what each of these mean for the data in real life. What does slope coef_ mean in this scenario, what do the x-intercepts represent. The slope means that for every dollar that is spent in producing the movie we should get around $3.1 revenue in return (Bigger budget films tend to do better). The intercept is telling us that a movie with a budget of 0 we would lose over 7 million dollars

```python
regression = LinearRegression()
regression.fit(x, y) 
print(regression.coef_) #this is the regression result
regression.intercept_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.3)
plt.plot(x, regression.predict(x), color="red", linewidth=4)

plt.title("Film Cost vs Global Rev")
plt.xlabel("Prod Budget")
plt.ylabel("Worldwide Revenue")
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show() 
```

This simple data can explain about 55 percent of the variation that we see in worldwide movies

```python
regression.score(x, y)
```
